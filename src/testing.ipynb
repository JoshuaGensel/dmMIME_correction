```python
# plot corrected Kds for all 3 first prot concentrations
mean_corrected_01 = np.mean([np.log(corrected_01_01), np.log(corrected_01_1), np.log(corrected_01_10)], axis=0)
mean_corrected_1 = np.mean([np.log(corrected_1_01), np.log(corrected_1_1), np.log(corrected_1_10)], axis=0)
mean_corrected_10 = np.mean([np.log(corrected_10_01), np.log(corrected_10_1), np.log(corrected_10_10)], axis=0)
plt.figure(figsize=(10, 10))
plt.scatter(np.log(true_01), mean_corrected_01, label='prot0.1', color='blue', alpha=0.5)
plt.scatter(np.log(true_01), mean_corrected_1, label='prot1', color='green', alpha=0.5)
plt.scatter(np.log(true_01), mean_corrected_10, label='prot10', color='red', alpha=0.5)
# also plot the average of the 3 corrected Kds
plt.scatter(np.log(true_01), np.mean([mean_corrected_01, mean_corrected_1, mean_corrected_10], axis=0), label='average', color='black', alpha=0.5, marker='x')
x = np.linspace(np.min(np.log(true_01)), np.max(np.log(true_01)), 100)
y = x
plt.plot(x, y, color='black', linestyle='--')
plt.legend()
plt.title('Corrected Kds for different first prot concentrations')
plt.show()

# plot avreage of the 3 corrected Kds vs the first round average Kds
plt.figure(figsize=(10, 10))
plt.scatter(np.log(true_01), np.mean([mean_corrected_01, mean_corrected_1, mean_corrected_10], axis=0), label='corrected 2nd round', color='red', alpha=0.5)
plt.scatter(np.log(true_01), np.mean([np.log(inferred_01), np.log(inferred_1), np.log(inferred_10)], axis=0), label='inferred 1st round', color='blue', alpha=0.5)
# also plot the average of the 3 uncorrected Kds
mean_uncorrected_01 = np.mean([np.log(inferred_01_01), np.log(inferred_01_1), np.log(inferred_01_10)], axis=0)
mean_uncorrected_1 = np.mean([np.log(inferred_1_01), np.log(inferred_1_1), np.log(inferred_1_10)], axis=0)
mean_uncorrected_10 = np.mean([np.log(inferred_10_01), np.log(inferred_10_1), np.log(inferred_10_10)], axis=0)
plt.scatter(np.log(true_01), np.mean([mean_uncorrected_01, mean_uncorrected_1, mean_uncorrected_10], axis=0), label='uncorrected 2nd round', color='green', alpha=0.5, marker='x')
x = np.linspace(np.min(np.log(true_01)), np.max(np.log(true_01)), 100)
y = x
plt.plot(x, y, color='black', linestyle='--')
plt.legend()
plt.title('Corrected vs first round Kds')
plt.show()
```
