```python
# plot corrected Kds for all 3 first prot concentrations
plt.figure(figsize=(10, 10))
plt.scatter(np.log(true_01), np.mean([np.log(corrected_01_01), np.log(corrected_01_1), np.log(corrected_01_10)], axis=0), label='prot0.1', color='blue', alpha=0.5)
plt.scatter(np.log(true_01), np.mean([np.log(corrected_1_01), np.log(corrected_1_1), np.log(corrected_1_10)], axis=0), label='prot1', color='green', alpha=0.5)
plt.scatter(np.log(true_01), np.mean([np.log(corrected_10_01), np.log(corrected_10_1), np.log(corrected_10_10)], axis=0), label='prot10', color='red', alpha=0.5)
x = np.linspace(np.min(np.log(true_01)), np.max(np.log(true_01)), 100)
y = x
plt.plot(x, y, color='black', linestyle='--')
plt.xlabel('True Kds')
plt.ylabel('Corrected Kds')
plt.legend()
plt.title('Corrected Kds for different first prot concentrations')
plt.show()
```
