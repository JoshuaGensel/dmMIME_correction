```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 5
theta = .2
# plot gamma distribution
samples = np.random.lognormal(1, 1, 100000)
plt.hist(samples, 50, density=True)
#add mean line
plt.axvline(np.mean(samples), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("mean: ", np.mean(samples))
```

```python
log_samples = np.log(samples)

# plot lognormal distribution
plt.hist(log_samples, 50, density=True)
#add mean line
plt.axvline(np.mean(log_samples), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
alpha = 4
beta = 2
# plot beta distribution
samples = np.exp((np.random.beta(alpha, beta, 1000000))*3)
plt.hist(samples, 100, density=True)
#add mean line
plt.axvline(np.mean(samples), color='r', linestyle='dashed', linewidth=2)
# # x axis limit
# plt.xlim(0, 10)
plt.show()
```

```python
log_samples = np.log(samples)
plt.hist(log_samples, 100, density=True)
#add mean line
plt.axvline(np.mean(log_samples), color='r', linestyle='dashed', linewidth=2)
```
