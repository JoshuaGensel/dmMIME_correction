```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 5
q = 4
```

```python
number_interactions = 0.5 * L * (q-1) * (L*q - L - 3)
interactions = np.arange(number_interactions)+1
print(number_interactions)
print(interactions)
```

```python
# generate interaction matrix
interaction_matrix = np.zeros((L*(q-1), L*(q-1)))

i = 0
for pos1 in range(L):
    for pos2 in range(pos1+1, L):
        for a1 in range(q-1):
            for a2 in range(q-1):
                interaction_matrix[pos1*(q-1)+a1, pos2*(q-1)+a2] = interactions[i]
                i += 1
# symmetrize
interaction_matrix += interaction_matrix.T 
#set diagonal to one
np.fill_diagonal(interaction_matrix, 1)

print(interaction_matrix)
```

```python
alpha = 4
beta = 2
# plot beta distribution
samples = np.exp((np.random.beta(alpha, beta, 1000000)-0)*3)
plt.hist(samples, 100, density=True)
#add mean line
plt.axvline(np.mean(samples), color='r', linestyle='dashed', linewidth=2)
# # x axis limit
# plt.xlim(0, 10)
plt.show()
```

```python
log_samples = np.log(samples)
plt.hist(log_samples, 100, density=True)
#add mean line
plt.axvline(np.mean(log_samples), color='r', linestyle='dashed', linewidth=2)
```

```python
# compute variance of log samples
print("mean: ", np.mean(log_samples))
print("variance: ", np.var(log_samples))
```
